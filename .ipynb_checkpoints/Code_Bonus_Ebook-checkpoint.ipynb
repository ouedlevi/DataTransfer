# Classification KNN avec visualisation de la frontière de décision

Auteur: Machine Learnia

Source : https://www.machinelearnia.com/apprendre-le-machine-learning-en-une-semaine

## Import des librairies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from matplotlib.colors import ListedColormap
```

## Création d'un jeu de données 2D avec 3 classes

```python
X, y = make_classification(
    n_samples=300,           # Nombre d'échantillons
    n_features=2,            # 2 dimensions pour la visualisation
    n_redundant=0,           # Pas de features redondantes
    n_informative=2,         # Toutes les features sont informatives
    n_clusters_per_class=1,  # Un cluster par classe
    n_classes=3,             # 3 classes différentes
    random_state=0
)

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Taille de l'ensemble d'entraînement: {X_train.shape[0]} échantillons")
print(f"Taille de l'ensemble de test: {X_test.shape[0]} échantillons")
print(f"Nombre de classes: {len(np.unique(y))}")
```

```python
## 2. Visualisation des données originales

plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green']
for i, color in enumerate(colors):
    plt.scatter(X[y == i, 0], X[y == i, 1],
               c=color, label=f'Classe {i}', alpha=0.7, s=50)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Distribution des données par classe')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Entraînement du modèle KNN

Comme vu dans le livre, l'algorithme de KNN dépend du nombre de voisin (k) que l'on choisir.

Le code ci-dessous présente donc les résultats pour différentes valeurs de k.

N'hésitez pas a jouer avec d'autres valeurs.


```python
# Test de différentes valeurs de k
k_values = [1, 3, 5, 7, 9, 15]
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"k={k}: Précision = {accuracy:.3f}")

# Sélection du meilleur k
best_k = k_values[np.argmax(accuracies)]
print(f"\nMeilleur k: {best_k} (précision: {max(accuracies):.3f})")

## 4. Visualisation des performances selon k

plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Nombre de voisins (k)')
plt.ylabel('Précision')
plt.title('Performance du KNN selon la valeur de k')
plt.grid(True, alpha=0.3)
plt.axvline(x=best_k, color='red', linestyle='--',
           label=f'Meilleur k = {best_k}')
plt.legend()
plt.show()
```

## visualiser la frontière de décision

dans le code ci-dessous, nous allons créer une fonction permettant de tracer les frontieres de décision (comme elles le sont présenter dans le livre que je vous ai offert).

Note : le code peut sembler impressionnant, mais il n'est pas si difficile a comprendre une fois que l'on maitrise les bases de **Numpy** et de **Matplotlib**

Voici 2 liens pour vous aider a prendre en main ces outils:

https://www.youtube.com/watch?v=O_OeWxpnUc0

https://www.youtube.com/watch?v=NzDQTrqsxas&t=1s


```python
def plot_decision_boundary(X, y, classifier, title="Frontière de décision KNN"):
    """
    Visualise la frontière de décision d'un classificateur
    """
    # Création d'une grille de points
    h = 0.02  # pas de la grille
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Prédiction sur tous les points de la grille
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Couleurs pour les régions et les points
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])
    cmap_bold = ['red', 'blue', 'green']

    plt.figure(figsize=(12, 8))

    # Affichage des régions de décision
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)

    # Affichage des points d'entraînement
    for i in range(len(np.unique(y))):
        plt.scatter(X[y == i, 0], X[y == i, 1],
                   c=cmap_bold[i], label=f'Classe {i}',
                   alpha=0.8, s=50, edgecolors='black')

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Visualisation pour k=1, k=5, et k=15
k_compare = [1, 5, 15]

for k in k_compare:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    plot_decision_boundary(X_train, y_train, knn,
                          f"Frontière de décision KNN (k={k})")

    # Évaluation sur l'ensemble de test
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nRésultats pour k={k}:")
    print(f"Précision: {accuracy:.3f}")
```

```python
## Modèle final avec le meilleur k

# Entraînement du modèle final
final_knn = KNeighborsClassifier(n_neighbors=best_k)
final_knn.fit(X_train, y_train)

# Prédictions finales
y_pred_final = final_knn.predict(X_test)

# Rapport de classification détaillé
print("\n" + "="*50)
print("RAPPORT FINAL")
print("="*50)
print(f"Modèle: KNN avec k={best_k}")
print(f"Précision globale: {accuracy_score(y_test, y_pred_final):.3f}")
print("\nRapport de classification détaillé:")
print(classification_report(y_test, y_pred_final))
```

## Visualisation finale avec les ensembles d'entraînement et de test

```python
plt.figure(figsize=(15, 5))

# Sous-graphique 1: Données d'entraînement avec frontière
plt.subplot(1, 3, 1)
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = final_knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])
cmap_bold = ['red', 'blue', 'green']

plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
for i in range(len(np.unique(y))):
    plt.scatter(X_train[y_train == i, 0], X_train[y_train == i, 1],
               c=cmap_bold[i], alpha=0.8, s=30, edgecolors='black')
plt.title(f'Entraînement (k={best_k})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Sous-graphique 2: Données de test
plt.subplot(1, 3, 2)
for i in range(len(np.unique(y))):
    plt.scatter(X_test[y_test == i, 0], X_test[y_test == i, 1],
               c=cmap_bold[i], alpha=0.8, s=30, label=f'Vraie classe {i}')
plt.title('Données de test (vraies classes)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# Sous-graphique 3: Prédictions sur les données de test
plt.subplot(1, 3, 3)
for i in range(len(np.unique(y))):
    plt.scatter(X_test[y_pred_final == i, 0], X_test[y_pred_final == i, 1],
               c=cmap_bold[i], alpha=0.8, s=30, label=f'Prédite classe {i}')
plt.title('Prédictions sur les données de test')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()

## 8. Analyse des erreurs de classification

# Identification des erreurs
errors = y_test != y_pred_final
error_indices = np.where(errors)[0]

if len(error_indices) > 0:
    print(f"\nNombre d'erreurs de classification: {len(error_indices)}")
    print("Exemples d'erreurs:")
    for i in error_indices[:5]:  # Afficher les 5 premières erreurs
        print(f"  Point {i}: Vraie classe = {y_test[i]}, "
              f"Prédite = {y_pred_final[i]}")

    # Visualisation des erreurs
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)

    # Points correctement classifiés
    correct = ~errors
    for i in range(len(np.unique(y))):
        mask = (y_test == i) & correct
        plt.scatter(X_test[mask, 0], X_test[mask, 1],
                   c=cmap_bold[i], alpha=0.7, s=50,
                   edgecolors='black', label=f'Classe {i} (correct)')

    # Points mal classifiés
    plt.scatter(X_test[errors, 0], X_test[errors, 1],
               c='yellow', s=100, marker='x', linewidth=3,
               label='Erreurs de classification')

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'Analyse des erreurs (k={best_k})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("\nAucune erreur de classification sur l'ensemble de test !")
```
